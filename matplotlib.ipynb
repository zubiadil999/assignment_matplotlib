```python
##Q2:Scatter plot: A scatter plot is used to display the relationship between two variables. It is often used to find patterns or correlations between two variables.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title ")
plt.show()
```

```python
##Q3:The subplot() function in Matplotlib is used to create multiple plots in a single figure. This function is particularly useful when you want to display multiple plots side by side to compare them or to present related information together.
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
fig, axs = plt.subplots(2, 2)

# Plotting line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plotting line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plotting line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plotting line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Displaying the plots
plt.show()
```

```python
##Q4:A bar plot is a type of chart that displays categorical data using rectangular bars. The height or length of each bar is proportional to the value it represents, and each bar represents a particular category or group. The bars can be vertical or horizontal, depending on the orientation of the chart.
##Bar plots are commonly used to compare data across different categories or groups. They are especially useful when you have discrete data, such as counts or frequencies, that you want to display visually. Bar plots can also be used to display continuous data by dividing it into discrete intervals or bins.
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("bar plot")
plt.show()
```

```python
##Q5:A box plot is a type of chart used to display the distribution of a dataset. It is also known as a box-and-whisker plot. The box in the plot represents the interquartile range (IQR) of the data, which is the range between the first and third quartiles. The median of the data is represented by a line inside the box. The whiskers extend from the box to show the range of the data outside the IQR, and any data points outside the whiskers are represented as outliers.

##Box plots are useful for comparing distributions of data across different categories or groups. They provide a visual summary of the central tendency, variability, and skewness of the data, and they can help identify potential outliers. Box plots are particularly useful when comparing data that are not normally distributed, or when comparing data with different sample sizes.
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
data = [box1, box2]
plt.boxplot(data)

# Adding labels and title
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box Plot')

# Displaying the plot
plt.show()
```
